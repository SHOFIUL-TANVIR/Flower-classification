```python
import math
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
a = [10,-4,2,6]
```

```python
def relu_activationFunction(number):
    numberStore = []
    for i in number:
        initialValue = 0
        if i > 0:
            initialValue = i
        numberStore.append(initialValue)
        
    return numberStore          
```

```python
randomNumber = np.arange(-20.0,20.0,0.2)
relu_line = relu_activationFunction (randomNumber)
plt.plot( randomNumber, relu_line)
ax = plt.gca()
plt.show()
```
