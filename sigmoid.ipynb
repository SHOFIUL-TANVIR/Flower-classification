```python
import math
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def sigmoidFunction(number):
    sigmoidDataStore = []
    
    for i in number:
        each_result = 1/(1+math.exp(-i))
        
        sigmoidDataStore.append(each_result)
    return sigmoidDataStore
```

```python
randomNumber = np.arange(-20.0,20.0,0.2)
sigmoid_plot = sigmoidFunction (randomNumber)
plt.plot( randomNumber, sigmoid_plot)
ax = plt.gca()
plt.show()
```
