```python
import math
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def tahhActivationFunctions(number):
    tanhValue = []
    for i in number:
        tanhValue.append(math.tanh(i))
    return tanhValue
```

```python
def tahhActivationFunctions(number):
    tanhValue = []
    for i in number:
        initialValue = 0
        if i < 0:
            initialValue = -1
            tanhValue.append(initialValue)

        elif i> 0:
            initialValue = 1
            tanhValue.append(initialValue)
        
    return tanhValue
```

```python
randomNumber = np.arange(-20.0,20.0,0.2)
tanhplot = tahhActivationFunctions(randomNumber)
plt.plot( randomNumber, tanhplot)
ax = plt.gca()
plt.show()
```
