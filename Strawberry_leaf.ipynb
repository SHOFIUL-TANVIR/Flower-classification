```python
import numpy as np
import os
import cv2
import random
import matplotlib.pyplot as plt
```

```python
DIRECTORY= "E:\PYTHON\Flower Classification\leaf"
CATEGORIES= ['Strawberry_fresh','Strawberry_scorch']
```

```python
data=[]

for categories in CATEGORIES:
    folder=os.path.join(DIRECTORY,categories)
    label=CATEGORIES.index(categories)
    
    
    for img in os.listdir(folder):
        img=os.path.join(folder,img)
        img_arr=cv2.imread(img)
        img_arr=cv2.resize(img_arr,(100,100))
        
        data.append([img_arr,label])
        
       
     
        
        
        
        
        
```

```python
data
```

```python
random.shuffle(data)
```

```python
x=[]
y=[]

for features,label in data:
    x.append(features)
    y.append(label)
```

```python
x= np.array(x)
y= np.array(y)
```

```python
x=x/255
```

```python
x
```

```python
x.shape
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D,MaxPooling2D,Flatten,Dense,Activation
```

```python
model=Sequential()
model.add(Conv2D(32,(3,3),input_shape=x.shape[1:],activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

model=Sequential()
model.add(Conv2D(32,(3,3),input_shape=x.shape[1:],activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Flatten())
model.add(Dense(2,activation='softmax'))
```

```python
model.compile(loss='sparse_categorical_crossentropy',optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(x,y,epochs=20,validation_split=0.1)
```

```python
from keras.preprocessing import image
import numpy as np

img_pred=image.load_img(r"E:\PYTHON\Flower Classification\leaf\Strawberry_scorch\1.jpg", target_size=(100,100))

img_pred=image.img_to_array(img_pred)
img_pred=np.expand_dims(img_pred, axis=0)

rslt=model.predict(img_pred)

print(rslt)
if rslt[0][0]>rslt[0][1]:
    prediction="Strawberry_fresh"
    
else:
    prediction="Strawberry_scorch"
print(prediction)
    
```
